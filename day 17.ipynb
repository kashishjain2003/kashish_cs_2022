```python
import matplotlib.pyplot as plt
import numpy as np
```

### types of graphs
1. plot -   line graph, between two parameters
2. bar -    bar graph of single parameter
3. hist -   histogram of a bunch of data i.e. frequency of single parameter
4. scatter -  dot graph between two parameters 
5. boxplot/box - statistical overview in form of box for a single param(optional-categorical data)
6. stem   -    more like bar graph,lines instead of bars
7. violinplot -   voilin shaped statistical overview graph
                  single parameter (optional- cat-data)


```python
x= np.linspace(0,5,11)
x1=np.arange(1,11,3)
```

```python
x
```

```python
y=x**3
```

```python
y2=x**4
```

```python
plt.plot(x,y)
#plt.show()
plt.title('plot graph')
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,lw=3,marker='o',markersize=15,markerfacecolor='green')
plt.plot(x,y2,lw=5,marker='v',markersize=10)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```

```python
fig.savefig('mygraph.png')
```

```python
plt.bar(x,y)
plt.title('bar graph')
```

```python
plt.hist(y)
plt.title('histogram')
```

```python
fig.savefig('mygraph.png')
```

```python
plt.scatter(x,y)
plt.title('scatter')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.boxplot(y)
plt.title('boxplot')
plt.grid()
```

```python
plt.stem(x,y)
plt.title('stem')
```

```python
plt.violinplot(x)
plt.title('violinplot')
```
