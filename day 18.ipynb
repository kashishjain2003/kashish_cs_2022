#### Graph Functions:
1. plot                    - Simple Line Plot
2. plot.area               - Filled version of Line Plot
3. plot.bar / plot.barh    - Bar Graph
4. plot.box                - box Plot
5. plot.density            - Density Line (Smooth Curved)
6. plot.hexbin             - Hexagon Version of scatter with density 
7. plot.hist               - Histogram
8. plot.kde                - kernel Density Estimator (density graph)
9. plot.pie                - Pie chart of categorical data
10. plot scatter           - Scatter graph

# Seaborn

```python
import seaborn as sns
```

##### Categories
1. Categorical plots
   
   a. countplot
   b. barplot (one categorical,one numerical, also calculates mean by default)
   c. boxplot
   d. violinplot
   e. swarmplot

2. Distribution Plots
   
   a. distplot
   b. displot
   c. jointplot
   d. pair plot
   e. kdeplot
   f. rugplot
   g. scatter plot 

3. Matrix plots
  
   a. heatmap
   b. clustermap
   c. pairplot

4. Regression plot
   
   a.lmplot
   b. regplot

5. Grid plots
   
   a. pairgrid
   b. jointgrid
   c. Facegrid

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe()
```

```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python



sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='time',estimator=np.mean)
```

```python
sns.boxplot(x='size',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker')
```
